```python
import pandas as pd
import re
csv_file_path = '../DATA/RAW/dirty_cafe_sales.csv'

csv_file = pd.read_csv(csv_file_path)
nb_column = len(csv_file.columns)
print(f'The number of columns is: {nb_column}')
print(f'Columns are: {csv_file.columns.tolist()}')
print(f'The number of rows is: {len(csv_file)}')
```

```python
dups_subset = csv_file[csv_file.duplicated(subset=['Transaction ID'], keep=False)]
print(f'There is {len(dups_subset)} duplicated rows')
```

```python
na_rows = csv_file[csv_file.isna().any(axis=1)]
print(f'There is {len(na_rows)} rows with missing values')
```

```python
wrong_txn_id = csv_file[~csv_file['Transaction ID'].str.fullmatch(r'TXN_\d{7}')]
print(f"There is {len(wrong_txn_id)} invalid values in 'Transaction ID' column")
```

```python
wrong_Item = csv_file[~csv_file['Item'].str.fullmatch(r'[A-Z][a-z]*', na=False)]

print(f"There is {len(wrong_Item)} invalid values in 'Item' column")
for item in wrong_Item['Item']:
    print(f'Invalid Item: {item}')
```

```python
wrong_Quantity = csv_file[~csv_file['Quantity'].str.fullmatch(r'\d+', na=False)]

print(f"There is {len(wrong_Quantity)} invalid values in 'Quantity' column")
for quantity in wrong_Quantity['Quantity']:
    print(f'Invalid Quantity: {quantity}')
```

```python
wrong_Price = csv_file[~csv_file['Price Per Unit'].str.fullmatch(r'\d+.\d', na=False)]

print(f"There is {len(wrong_Price)} invalid values in 'Price Per Unit' column")
for price in wrong_Price['Price Per Unit']:
    print(f'Invalid Price: {price}')
```

```python
wrong_Total_Spent = csv_file[~csv_file['Total Spent'].str.fullmatch(r'\d+.\d', na=False)]

print(f"There is {len(wrong_Total_Spent)} invalid values in 'Total Spent' column")
for total_spent in wrong_Total_Spent['Total Spent']:
    print(f'Invalid Total Spent: {total_spent}')
```

```python
wrong_Payment_Method = csv_file[~csv_file['Payment Method'].str.fullmatch(r'([A-Z][a-z]* ?)+', na=False)]

print(f"There is {len(wrong_Payment_Method)} invalid values in 'Payment Method' column")
for payment_method in wrong_Payment_Method['Payment Method']:
    print(f'Invalid Payment Method: {payment_method}')
```

```python
wrong_Location = csv_file[~csv_file['Location'].str.fullmatch(r'(In-store)|(Takeaway)', na=False)]

print(f"There is {len(wrong_Location)} invalid values in 'Location' column")
for location in wrong_Location['Location']:
    print(f'Invalid Location: {location}')
```

```python
wrong_Date = csv_file[~csv_file['Transaction Date'].str.fullmatch(r'\d{4}-\d{2}-\d{2}', na=False)]

print(f"There is {len(wrong_Date)} invalid values in 'Transaction Date' column")
for date in wrong_Date['Transaction Date']:
    print(f'Invalid Transaction Date: {date}')
```

```python
#Columns are: ['Transaction ID', 'Item', 'Quantity', 'Price Per Unit', 'Total Spent', 'Payment Method', 'Location', 'Transaction Date']

error_dfs = {
    "Transaction ID": wrong_txn_id,
    "Item": wrong_Item,
    "Quantity": wrong_Quantity,
    "Price Per Unit": wrong_Price,
    "Total Spent": wrong_Total_Spent,
    "Payment Method": wrong_Payment_Method,
    "Location": wrong_Location,
    "Transaction Date": wrong_Date,
}

error_summary = {
    col: err_df[col].value_counts(dropna=False).to_dict()
    for col, err_df in error_dfs.items()
}
print("Error Summary:")
for col, summary in error_summary.items():
    print(f"  {col}: {summary}")
```
