```python
import pandas as pd

csv_file_path = '../DATA/PROCESSED/cleaned_cafe_sales.csv'

csv_file = pd.read_csv(csv_file_path)
```

```python
# Convert 'Transaction Date' to datetime
csv_file['Transaction Date'] = pd.to_datetime(csv_file['Transaction Date'])

#Add additional temporal columns
csv_file['Year'] = csv_file['Transaction Date'].dt.year
csv_file['Month'] = csv_file['Transaction Date'].dt.month
csv_file['Day'] = csv_file['Transaction Date'].dt.day
csv_file['Weekday'] = csv_file['Transaction Date'].dt.day_name()
```

```python
revenue_by_item = (
    csv_file
    .groupby('Item', dropna=False)
    .agg(
        price=('Price Per Unit', 'mean'),
        total_revenue=('Total Spent', 'sum'),
        total_quantity=('Quantity', 'sum'),
        nb_transactions=('Transaction ID', 'count')
    )
    .sort_values('total_revenue', ascending=False)
)
print(revenue_by_item)

avg_ticket_by_item = (
    csv_file
    .groupby('Item')
    .agg(
        price=('Price Per Unit', 'mean'),
        avg_quantity=('Quantity', 'mean'),
        avg_ticket=('Total Spent', 'mean'))
).sort_values('avg_ticket', ascending=False)
print(avg_ticket_by_item)
```
